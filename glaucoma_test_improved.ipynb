```python
import cv2


def denoise_image(img):
   
    img_denoised = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)
    return img_denoised  


if 'images' in locals() or 'images' in globals():
    denoised_images = []
    for i, img in enumerate(images):
        print(f"Processing image {i+1}/{len(images)}...")
        denoised_images.append(denoise_image(img))
    print("Denoising completed for all images.")
else:
    print("Error: 'images' is not defined. Ensure images are loaded before running.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def edge_detection_segmentation(img):
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian Blur to smooth image and reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Canny Edge Detection
    edges = cv2.Canny(blurred, threshold1=50, threshold2=150)

    # Find contours from edges
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the original image
    segmented_img = img.copy()
    cv2.drawContours(segmented_img, contours, -1, (0, 255, 0), 2)  # Green contours

    return segmented_img, edges

# Apply edge detection to all images
segmented_images = []
edge_maps = []

for img in images:
    segmented_img, edges = edge_detection_segmentation(img)
    segmented_images.append(segmented_img)
    edge_maps.append(edges)

print("Edge detection segmentation completed!")

# Display an example result
plt.figure(figsize=(10, 5))

# Show Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(images[0], cv2.COLOR_BGR2RGB))
plt.title("Original Image")

# Show Edge Map
plt.subplot(1, 3, 2)
plt.imshow(edge_maps[0], cmap="gray")
plt.title("Canny Edge Map")

# Show Segmented Image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(segmented_images[0], cv2.COLOR_BGR2RGB))
plt.title("Segmented Image with Contours")



plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Load MobileNetV2 with pretrained weights
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True  # Unfreeze some layers for better feature extraction

# Fine-tune only the last 80 layers
for layer in base_model.layers[:-80]:
    layer.trainable = False

# Add custom layers on top
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)  # Normalize activations
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)  # Reduce overfitting
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)  # Additional normalization
x = Dropout(0.4)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(8, activation='softmax')(x)  # Multi-class classification

# Define the final model
model = Model(inputs=base_model.input, outputs=x)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.00005), loss='categorical_crossentropy', metrics=['accuracy'])

# Data directory (combined dataset)
data_dir = 'C:/Users/KIIT/Desktop/datasets/G1020'

# Verify dataset structure
if not os.path.exists(data_dir):
    raise ValueError("Dataset directory not found. Please check the path.")

# Data augmentation and preprocessing with validation split
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.3,
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.3,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    validation_split=0.2  # 80% training, 20% validation
)

# Load training and validation datasets
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',  # Multi-class classification
    subset='training'
)
print(f"Training samples found: {train_generator.samples}")

validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',  # Multi-class classification
    subset='validation'
)
print(f"Validation samples found: {validation_generator.samples}")

# Ensure dataset is not empty
if train_generator.samples == 0 or validation_generator.samples == 0:
    raise ValueError("No images found in dataset directory. Ensure proper structure and file formats.")

# Train the model
epochs = 10
model.fit(train_generator, validation_data=validation_generator, epochs=epochs)

# Save model
model.save('mobilenetv2_glaucoma.h5')

# Evaluate the model
loss, accuracy = model.evaluate(validation_generator)
print(f"Validation Accuracy: {accuracy * 100:.2f}%")
```
